## Simple line plots

```python
x = [-1, 3, 4, 8, 10]
f = [-1, -2, 7, 13, 1]
```

```python
plt.plot(x, f);
```

```python
plt.plot(x, f, '-or')
plt.xlabel('$x$')
plt.ylabel('$f$')
plt.title("Simple plot of $x$ vs $f$");
```

## Plotting mathematical functions of one variable


```python
num_points = 100
x = np.linspace(0, 4*np.pi, num=num_points)
```

```python
x
```

```python
f = np.sin(x)
```

```python
plt.plot(x, f)
plt.xlabel('$x$')
plt.ylabel('$\sin(x)$')
plt.xlim(x[0], x[-1]);
```

```python
# To plot two functions in one graph, we use plt.plot twice:

plt.plot(x, np.sin(x), label='$\sin(x)$')
plt.plot(x, np.cos(x), label='$\cos(x)$')
plt.xlabel('$x$')
plt.legend()
plt.xlim([x[0], x[-1]]);
```

# Bar charts and histograms

## Bar charts

```python
cohort = ('IA', 'IB', 'IIA', 'IIB', 'Graduate')
num_students = (332, 270, 256, 238, 451)
```

```python
x_pos = np.arange(len(cohort))
```

```python
plt.bar(x_pos, num_students, align='center')
plt.xticks(x_pos, cohort, rotation=30)
plt.xlabel('cohort')
plt.ylabel('number of students');
```

## Histograms

```python
x = np.random.normal(loc=0.0, scale=1.0, size=5000)
```

```python
x
```

```python
n, bins, patches = plt.hist(x, 20, facecolor='green')
plt.xlabel('value')
plt.ylabel('frequency');
```

```python
print(bins)
```

```python
print(n)
```

```python
from scipy.stats import norm

n, bins, patches = plt.hist(x, 20, normed=1.0, facecolor='green', alpha=0.75, edgecolor='black')
xf = np.linspace(bins[0], bins[-1], 100)
f = norm.pdf(xf)
plt.plot(xf, f, 'b')
plt.xlabel('value')
plt.ylabel('frequency');
```

## Interactive plots

```python
from ipywidgets import interact
```

```python
def plot(X=1, Y=0):
    t = np.linspace(0, 2*np.pi, 200)
    plt.xlabel('$t$')
    plt.ylabel('$f$')
    plt.title(r"$\alpha$ = {}, $\omega$ = {}".format(Y, X))
    plt.plot(t, (t**Y)*np.sin(X*t))
```

```python
interact(plot, X=(-10, 10, 0.25), Y=(0, 2, 0.25));
```
